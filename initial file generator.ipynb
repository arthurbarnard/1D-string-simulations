```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
Npoints = 100
Length = 10000
epsilon = .995
_dX = Length/Npoints
_x = np.linspace(0,Length*epsilon,Npoints)[:]
_z = np.zeros(np.shape(_x))
_n1 = int(np.floor(Npoints/2))
_z[0:_n1] = -_x[0:_n1]*np.arccos(epsilon)
_n2 = int(np.ceil(Npoints/2))
_z[_n2:] = -((Length)*epsilon-_x[_n2:])*np.arccos(epsilon)

x = np.zeros(Npoints+2)
x[1:-1] = _x +_dX
x[-1] = _x[-1]+2*_dX

z = np.zeros(Npoints+2)
z[1:-1] = _z

_outdata = np.zeros((Npoints+2)*6)

_outdata[0::6] = x
_outdata[2::6] = z

plt.plot(x,z)
plt.xlabel('x-position (nm)')
plt.ylabel('z-position (nm)')

plt.show()
```

```python
#generating output file
_fnum = 0
_fname = './initfiles/triangle_test_%03d.txt' % _fnum

_fid = open(_fname,'w+b')
_fid.write(bytearray(_outdata))
_fid.close()
```

```python
#loading in initial file

_fid = open(_fname)
_data = np.fromfile(_fid,dtype = 'float')
N_points = len(_data)/6
_fid.close()
x_init = _data[0::6]
z_init = _data[2::6]

#plot initial file
plt.plot(x_init,z_init)
plt.xlabel('x-position (nm)')
plt.ylabel('y-position (nm)')
plt.show()
```
